```python
import numpy as np
from sklearn.feature_selection import VarianceThreshold
```

```python
a = np.random.randn(100)
np.var(a)
```

```python
b = np.random.randn(100) * 0.1
np.var(b)
```

```python
X = np.vstack([a,b]).T
X
```

```python
# 低方差过滤
vt = VarianceThreshold(0.02)
vt.fit_transform(X)
```
